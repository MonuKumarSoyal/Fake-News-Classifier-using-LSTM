```python
import pandas as pd
```

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

```python
dataset = dataset.dropna()
```

```python
dataset.head(10)
```

```python
print(dataset.shape)
dataset.reset_index(inplace = True)
title = dataset["title"]
title.head()
```

```python
import nltk
import re
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
title_list = []
stemmer = PorterStemmer()

for i in range(title.shape[0]):
    temp = re.sub('[^a-zA-Z]', " ", title[i])
    temp = re.sub('\s+', ' ', temp)
    temp = temp.lower()
    temp = temp.split()
    temp = [stemmer.stem(words) for words in temp if words not in set(stopwords.words('english'))]
    temp = ' '.join(temp)
    title_list.append(temp)

title_list
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
title_list_1 = []
stemmer_1 = PorterStemmer()
sum = 0

for i in range(title.shape[0]):
    temp = re.sub('[^a-zA-Z]', " ", title[i])
    temp = re.sub('\s+', ' ', temp)
    temp = temp.lower()
    temp = temp.split()
    temp = [words for words in temp if words not in set(stopwords.words('english'))]
    sum = sum + len(temp)
```

```python
sum # number of total words present in title after text preprocessing
```

```python
voc_size = 10000
one_hot_repre = [one_hot(words, voc_size) for words in title_list]
one_hot_repre
```

```python
input = pad_sequences(one_hot_repre, padding = "pre", maxlen = 20)
input
```

```python
model = Sequential()
model.add(Embedding(voc_size, 40, input_length = 20))
model.add(LSTM(100))
model.add(Dense(1, activation = "sigmoid"))
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
print(model.summary())
```

```python
input[0]
```

```python
y = dataset["label"]
```

```python
import numpy as np
input = np.array(input)
y = np.array(y)
```

```python
input.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input, y, test_size = 0.33, random_state = 0)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model.fit(x_train, y_train, validation_data =(x_test, y_test), epochs = 10, batch_size = 64)
```

```python
y_pred = model.predict(x_test)
```

```python
for i in range(y_pred.shape[0]):
    if(y_pred[i] < 0.5):
        y_pred[i] = 0
    else:
        y_pred[i] = 1
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
matrix = confusion_matrix(y_test, y_pred)
matrix
```

```python
score = accuracy_score(y_test, y_pred)
score
```

### Adding Dropout ratio

```python
from tensorflow.keras.layers import Dropout
feature_vector = 40
model_1 = Sequential()
model_1.add(Embedding(voc_size, feature_vector, input_length = 20))
model_1.add(Dropout(0.3))
model_1.add(LSTM(100))
model_1.add(Dropout(0.3))
model_1.add(Dense(1, activation = "sigmoid"))
model_1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model_1.summary())
```

```python
model_1.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 10, batch_size = 64)
```

```python
y_pred_1 = model_1.predict(x_test)

for i in range(y_pred_1.shape[0]):
    if(y_pred_1[i] < 0.5):
        y_pred_1[i] = 0
    else:
        y_pred_1[i] = 1
```
